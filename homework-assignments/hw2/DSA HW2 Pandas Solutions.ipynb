#  Задание 12
Найдите среднюю температуру в городах стран, входящих в Европейский союз, и не входящих в Европейский Союз ДО и ПОСЛЕ "Брексита". То есть, ваш ответ должен быть в формате:

Средняя температура по EU до Brexit: ...
Средняя температура по non-EU до Brexit: ...

Средняя температура по EU после Brexit: ...
Средняя температура по non-EU после Brexit: ...

То есть для вычисления значений первой пары Великобритания (United Kingdom) учитывается как EU='yes', а для вычисления значений второй пары, вы не учитываете Великобританию в группе EU (притворяетесь, будто она EU='no').

Подсказка 1: Что бы вы не делали, в любом случае, сначала объедините таблицы cities и countries
Подсказка 2: эту задачу можно решить 2-мя способами:

1 способ: создать две копии результата объединения двух таблиц. В одной оставить EU='yes' для United Kingdom.
Во второй заменить для нее значения с 'yes' на 'no'. Сгруппировать по EU/non-EU и рассчитать mean() температуры.

2 способ: Группировать по EU/non-EU ничего не надо. Вместо группировки вычислите каждую из 4 температур в отдельно, используя 4 двойных разных условия (вспомните про комбинацию условий посредством И (&), ИЛИ (|) и НЕ (~)).

```python
# Вариант #1

df_before = pd.merge(cities, countries, on='country')

print("Avg temperature of EU countries before Brexit: %0.3f"%df_before.temperature[df_before.EU=='yes'].mean())
print("Avg temperature of non-EU countries before Brexit: %0.3f"%df_before.temperature[df_before.EU=='no'].mean())

df_after = df_before.copy()
df_after.loc[df_after.country=='United Kingdom', 'EU'] = 'no'

print("Avg temperature of EU countries after Brexit: %0.3f"%df_after.temperature[df_after.EU=='yes'].mean())
print("Avg temperature of non-EU countries after Brexit: %0.3f"%df_after.temperature[df_after.EU=='no'].mean())
```

```python
# Вариант #2
df = pd.merge(cities, countries, on='country')
print("Avg temperature of EU countries before Brexit: %0.3f"%df.temperature[df.EU=='yes'].mean())
print("Avg temperature of non-EU countries before Brexit: %0.3f"%df.temperature[df.EU=='no'].mean())

print("Avg temperature of EU countries after Brexit: %0.3f"\
      %df.temperature[(df.EU=='yes') & (~(df.country=='United Kingdom'))].mean())
print("Avg temperature of non-EU countries after Brexit: %0.3f"\
      %df.temperature[(df.EU=='no') | (df.country=='United Kingdom')].mean())
```

```python
# Вариант #3

def beforeBrexit(row):
    if row['country']=='United Kingdom':
        row['EU'] = 'no'
    return row

result = pd.merge(cities, countries, how='inner', on='country')
before = result.groupby('EU').mean()['temperature'] #after brexit
after = result.apply(beforeBrexit,axis=1).groupby('EU').mean()['temperature']
result = pd.DataFrame([before,after],index=['Before Brexit','After Brexit'])
result
```
